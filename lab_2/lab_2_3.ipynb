```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
import csv
```

```python
data = []
with open('dataset_home.txt', 'r') as file:
    csv_reader = csv.reader(file)
    for row in csv_reader:
        data.append(row)
age = []
sex = []
bmi = []
children = []
smoker = []
region = []
charges = []
for row in data:
    age.append(int(row[0]))
    sex.append(row[1])
    bmi.append(float(row[2]))
    children.append(int(row[3]))
    smoker.append(row[4])
    region.append(row[5])
    charges.append(float(row[6]))
```

```python
# Scatter plot для Возраста и Расходов
plt.scatter(age, charges, alpha=0.5)
plt.title('Возраст vs Расходы')
plt.xlabel('Возраст')
plt.ylabel('Расходы')
plt.show()
```

Исходя из результатов, показанных на графике, можно заметить корреляцию между тем, как возраст влияет на расходы. Можно предположить, что с возрастом расходы увеличиваются.

```python
# Scatter plot для Возраста и количества детей
plt.scatter(age, children, alpha=0.5)
plt.title('Scatter Plot: Возраст vs Дети')
plt.xlabel('Возраст')
plt.ylabel('Дети')
plt.show()
```

Исходя из результатов, показанных на графике, можно заметить, что в наборе данных типичные значения для показателей возраста и количества детей. Больше всего, 5 детей у людей возраста до 50 лет. Как и стоило ожидать, больше всего у опрошенных, от 1 до 3 детей. Однако, согласно результатам, у большого количество людей нет детей.  

```python
# Линейный график для Расходов
plt.plot(charges)
plt.title('Line Plot: Расходы vs Количество опрошенных')
plt.xlabel('Количество опрошенных')
plt.ylabel('Расходы')
plt.show()
```

Как можно увидеть из графика, количество людей, расходующих большую сумму денег, от 50.000 до 60.000 – единицы. В основном, люди тратят менее 20.000 – 30.000.



```python
# Гистограмма для Возраста
plt.hist(age, bins=80, edgecolor='black')
plt.title('Histogram: Возраст')
plt.xlabel('Возраст')
plt.ylabel('Опрошенные люди')
plt.show()
```

Из графика видно, что большинство опрошенных - люди до 20 лет и количество опрошенных более старшего возраста - уменьшается.




```python
# Bar chart для Smoker
unique_smokers, counts = np.unique(smoker, return_counts=True)
plt.bar(unique_smokers, counts, color=['blue', 'red'])
plt.title('Bar Chart: Курящих людей')
plt.xlabel('Курящих людей')
plt.ylabel('Количество опрошенных')
plt.show()
```

По графику видно, большинство опрошенных не курят, что очень сильно радует.

```python
# Визуализация корреляций с использованием тепловой карты
correlation_matrix = np.corrcoef([age, bmi, charges])
plt.figure(figsize=(8, 6))
heatmap = plt.matshow(correlation_matrix, cmap="coolwarm")
plt.colorbar(heatmap, label="Correlation Coefficient")
plt.title("Correlation Heatmap")
plt.show()
# Рассчет корреляции между Age и Charges
correlation, _ = pearsonr(age, charges)
print(f'Correlation between Age and Charges: {correlation}')
# Рассчет корреляции между BMI и Charges
correlation, _ = pearsonr(bmi, charges)
print(f'Correlation between BMI and Charges: {correlation}')
```

Тепловая карта показывает взаимосвязь между возрастом, индексом массы тела и расходов. Темные ячейки показывают более сильную кореляцию.

```python
# Гистограмма для BMI
plt.hist(bmi, bins=40, edgecolor='black')
plt.title('Histogram: BMI')
plt.xlabel('ИМТ')
plt.ylabel('Опрошенные люди')
plt.show()
```

Как видно из гистограммы, большинство опрошенных людей имеют ИМТ от 25 до 35. Что является признаком избыточной массы тела или даже ожирение первой степени. Нормальные показатели ИМТ – от 18,5 до 25. Индекс Массы Тела выше 40 (что соответствует ожирению третьей степени) имеют единицы.
